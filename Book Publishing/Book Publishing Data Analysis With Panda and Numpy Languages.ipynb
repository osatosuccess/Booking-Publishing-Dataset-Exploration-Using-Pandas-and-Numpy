```python
extr = Book['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
Book['Date of Publication'] = pd.to_numeric(extr)
Book['Date of Publication'].dtype
```

```python
Book['Date of Publication'].isnull().sum()/ len(Book)
```

```python
Book['Place of Publication'].head(10)
```

```python
publication = Book['Place of Publication']
london = publication.str.contains('london')
london[:5]
```

```python
oxford = publication.str.contains('oxford')
```

```python
Book['Place of Publication'] = np.where(london, 'london',
                                       np.where(oxford, 'oxford',
                                                publication.str.replace('-', ' ')))
```
